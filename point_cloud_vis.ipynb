```python
import pypcd
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

path = 'data/pcd/trainEnv_9_362644000.pcd'
pc = pypcd.PointCloud.from_path(path)
# flatten into vector
# obs_pc = np.zeros((3, pc.pc_data['x'].shape[0]))
obs_pc = np.zeros((3, len(pc.pc_data['x'])))
obs_pc[0] = pc.pc_data['x']
obs_pc[1] = pc.pc_data['y']
obs_pc[2] = pc.pc_data['z']

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(obs_pc[0], obs_pc[1], obs_pc[2])
plt.show()
```
